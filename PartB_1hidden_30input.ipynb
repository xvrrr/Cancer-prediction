```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
###################
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers import Activation
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from keras.utils import to_categorical
###################


# Loading data from the .csv file
data_raw = pd.read_csv('BCW_dataset.csv', delimiter=',', header=0, index_col=None)
# Head method show first 5 rows of data
print(data_raw.head())




# Drop unused columns
drop_columns = ['Unnamed: 32', 'id', 'diagnosis']
# Convert strings -> integers
d = {'M': 0, 'B': 1}
# Define features and labels
y = data_raw['diagnosis'].map(d)
X = data_raw.drop(drop_columns, axis=1)
print('X.shape: ',X.shape)
print('y.shape: ',y.shape)
# testing the data values
print(y.values )
y_tmp=y.values
print(y_tmp[0:22])






import seaborn as sns
import matplotlib.pyplot as plt
# Split dataset into training (75%) and test (25%) set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
# Normalization of the data
X_train_n = (X_train-X_train.mean())/(X_train.max()-X_train.min())
X_test_n = (X_test-X_test.mean())/(X_test.max()-X_test.min())
print(X_train_n.shape)
print(y_train.shape)
print(y[0:3])
#print(X_train_n)


# fix random seed for reproducibility


#y_train = to_categorical(y_train)
#y_test = to_categorical(y_test)
num_classes = 2
num_hidden_nodes = 650

```

# New Section

```python
# ANN example
#




#
# one hot encode outputs


# define baseline model
# Dense : a regular fully connected layer
# The chosen complie model has a loss function, and the purpose is to compute the quantity that
# the model should minimize during training.  Below, the 'categorical_crossentropy' is used as
# the loss function. (The MSE is not chosen to be the loss function.) It would compute the crossentropy
# loss between the labels and predictions. There are two or more label classes, usually with labels
# provided in a one_hot representation.

# Removed one-hot encoding for the labels y_train and y_test



def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(num_hidden_nodes, input_dim=30, kernel_initializer='normal', activation='relu'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    # Choice of optimizer: adam (adaptive moment estimation), AdaGrad (adaptive learning rate),
    # sgd (Stochastic gradient descent), RMSprop (similar to AdaGrad), Adadelta (adaptive delta) ...
    return model
# build the model
model = baseline_model()

# Fit the model
# verbose : number of steps before showing a result
history = model.fit(X_train_n, y_train, validation_data=(X_test_n, y_test), epochs=25, batch_size=200, verbose=2)
# Final evaluation of the model
scores = model.evaluate(X_test_n, y_test, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores[1]*100))
#
metrics = history.history
plt.plot(history.epoch, metrics['accuracy'], metrics['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])
plt.show()

plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()


#Y_train = to_categorical(y_train)
#Y_test = to_categorical(y_test)


#y_pred = model.predict(X_test_n)
#y_pred_class = np.argmax(y_pred, axis=1)
#y_test = np.argmax(y_test, axis=1)
print(y_test)
print('hi')

#print(y_pred_class)
#print(y_test_class)
#print(y_pred_class.shape)
#print(y_test_class.shape)

# Confusion matrix

y_pred = model.predict(X_test_n)
y_pred = np.argmax(y_pred, axis=1)
print(y_test)
print(y_pred)


C=confusion_matrix(y_test, y_pred,labels=[0,1])
print(C)

sns.set()
f,ax = plt.subplots()
sns.heatmap(C,annot=True,ax=ax)

ax.set_title('confusion matrix')
ax.set_xlabel('predict')
ax.set_ylabel('true')



# Calculate precision
prec = precision_score(y_test, y_pred)
print("Precision:", prec)

# Calculate recall
rec = recall_score(y_test, y_pred)
print("Recall:", rec)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)
print("F1 score:", f1)
```
